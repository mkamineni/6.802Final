This file creates images containing RNA expression data for many patient samples and genes. The methods used to create these images are in https://journals.plos.org/plosone/article/file?id=10.1371/journal.pone.0230536&type=printable. The tree map algorithm reference can be found in  https://plotly.com/python/treemaps/. 

```python
import pandas as pd
import numpy as np
import json
import plotly.express as px
```

```python
df = pd.read_csv('norm_data.tsv', sep = '\t')
df
```

```python
f = open("GeneNames.txt", "r")
lines = [line.strip() for line in f.readlines()]
df['GeneName'] = lines
df
```

```python
f = open("briteId_to_geneId.txt", "r")
lines = f.readlines()
briteIDs = []
geneIDs = []
for elem in lines:
    if elem:
        ids = elem.split()
        briteIDs.append(ids[0])
        geneIDs.append(ids[1])
geneID_to_briteID = dict(zip(geneIDs, briteIDs))
```

```python
f = open("geneId_to_geneName.txt", "r")
lines = f.readlines()
geneIDs = []
geneNames = []
for elem in lines:
    if elem:
        ids = elem.split()
        geneIDs.extend([ids[0]]*(len(ids)-1))
        geneNames.extend(ids[1:])
geneNames = [elem.strip(';') for elem in geneNames]
geneNames = [elem.strip(',') for elem in geneNames]
geneName_to_geneID = dict(zip(geneNames, geneIDs))
print(len(geneName_to_geneID))
```

```python
f = open('geneNames.txt', 'r')
lines = [elem.strip() for elem in f.readlines()]
print(len(lines))
#get briteID for each element
briteIDs = []
gene_inds_to_keep = []
for ind in range(len(lines)):
    elem = lines[ind]
    if elem in geneName_to_geneID:
        geneID = geneName_to_geneID[elem]
        if geneID in geneID_to_briteID:
            briteIDs.append(geneID_to_briteID[geneID])
            gene_inds_to_keep.append(ind)
print("The number of genes that mappped to functional hierarchies: %d" %len(briteIDs))
print("The number of different functional hierarchies available: %d" %len(set(briteIDs)))
briteIDs = ['br'+elem[-5:] for elem in briteIDs]
f = open("briteIDs.txt", "w")
f.write(str(briteIDs))
f.close()
briteIDs
```

```python
df = df.loc[gene_inds_to_keep]
df
```

```python
df['keggBriteId'] = [elem[-4:] for elem in briteIDs]
df
```

```python
brite = pd.read_csv('KEGG_brite.csv')
gene_mapping = pd.merge(df, brite, on='keggBriteId')
gene_mapping
```

```python
#Number of genes present in final sample
len(set(gene_mapping['GeneName']))
```

```python
f = open('classifications.txt', 'r').read()
samples_to_label = eval(f)
print(len(samples_to_label))
samples = [col for col in df.columns if 'SR' in col]
crohns = []
colitis = []
for ind in range(len(samples)):
    classification = samples_to_label[samples[ind]]
    if classification == 0:
        crohns.append(ind)
    elif classification == 1:
        colitis.append(ind)
        
print(crohns)
```

```python
#need to filter out genes that are not expressed a lot
gene_inds_to_keep = []
expression_vals = gene_mapping.to_numpy()
print(expression_vals.shape)
for ind in range(expression_vals.shape[0]):
    gene_expr = expression_vals[ind]
    crohns_mean = np.array([gene_expr[i] for i in crohns]).mean()
    crohns_var = np.array([gene_expr[i] for i in crohns]).var()
    colitis_mean = np.array([gene_expr[i] for i in colitis]).mean()
    colitis_var = np.array([gene_expr[i] for i in colitis]).var()
    Z = (crohns_mean - colitis_mean) / np.sqrt(crohns_var/len(crohns) + colitis_var/len(colitis))
    if abs(Z) > 1.0023018:
        gene_inds_to_keep.append(ind)
len(gene_inds_to_keep)
```

```python
small_gene_mapping = gene_mapping.iloc[gene_inds_to_keep]
```
